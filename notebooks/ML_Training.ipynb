```python
import seaborn as sb
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Finalized_Dataset.csv")
df.head()
```

```python
x=df.drop("FWI",axis=1)
y=df[["FWI"]]
```

```python
x.head()
```

```python
y.head()
```

```python
x.drop(['year','month','day'],inplace=True,axis=1)
```

```python
data=x.corr()
data
```

```python
def feature_selection(dataset,threshold):
    col=set()
    for i in range(len(dataset.columns)):
        for j in range(i):
            if (abs(dataset.iloc[i,j])>=threshold):
                col.add(dataset.columns[i])
    return col

#print(feature_selection(data,0.85))

x.drop(feature_selection(data,0.85),axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
z=scaler.fit_transform(x_train)
test_data=scaler.transform(x_test)
```

```python
for i in range(len(x_train)):
    x_train.iloc[i]=z[i]
for i in range(len(x_test)):
    x_test.iloc[i]=test_data[i]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Ridge

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
linear_reg=LinearRegression()
linear_reg.fit(x_train,y_train)
error=mean_absolute_error(y_test,linear_reg.predict(x_test))
r2=r2_score(y_test,linear_reg.predict(x_test))
print(error)
print(r2*100)
```

```python
lasso_reg=Lasso()
lasso_reg.fit(x_train,y_train)
error=mean_absolute_error(y_test,lasso_reg.predict(x_test))
r2=r2_score(y_test,lasso_reg.predict(x_test))
print(error)
print(r2*100)
```

```python
ridge=Ridge()
ridge.fit(x_train,y_train)
error=mean_absolute_error(y_test,ridge.predict(x_test))
r2=r2_score(y_test,ridge.predict(x_test))
print(error)
print(r2*100)
```

```python
en=ElasticNet()
en.fit(x_train,y_train)
error=mean_absolute_error(y_test,en.predict(x_test))
r2=r2_score(y_test,en.predict(x_test))
print(error)
print(r2*100)
```

```python
import pickle

with open("scaler.pkl","wb") as file:
    pickle.dump(scaler,file)
with open("ridge_model.pkl","wb") as file:
    pickle.dump(ridge,file)
```
